```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Time (0 to 60 seconds)
t = np.linspace(0, 60, 100)

# 2. The "Understanding" Curve
# We'll use a 'Sigmoid' shape: slow start, fast 'Aha!' moment, then leveling off.
understanding = 1 / (1 + np.exp(-(t-30)/5))

# 3. The Derivative (Calculus in action!)
# This calculates the 'Slope' at every single point.
learning_speed = np.gradient(understanding, t)

# 4. Plotting
plt.figure(figsize=(10, 5))

# Plot the state of knowledge
plt.plot(t, understanding, label='Total Understanding (0 to 1)', color='blue', lw=3)

# Plot the speed of learning
plt.plot(t, learning_speed * 10, label='Learning Speed (Scaled x10)', color='orange', linestyle='--')

plt.title("Calculus in the OR: When does the 'Aha!' moment happen?")
plt.xlabel("Seconds into Briefing")
plt.ylabel("Intensity")
plt.axvline(30, color='red', alpha=0.3, label='Peak Learning Point')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Setup: 50 Patients, 1000ms each
n_patients = 50
duration = 1000
all_spikes = np.zeros((n_patients, duration))

# 2. Simulate the "Empathy Growth" for every patient
for i in range(n_patients):
    # Each patient has a slightly different 'Aha!' moment (Noise!)
    prob = np.linspace(0, 0.4, duration) + np.random.normal(0, 0.05, duration)
    prob = np.clip(prob, 0, 1) # Keep probabilities between 0 and 1
    all_spikes[i, :] = np.random.random(duration) < prob

# 3. Create the PSTH (The Average)
# We sum all spikes at each millisecond and divide by the number of patients
psth = np.mean(all_spikes, axis=0) * 1000 # Convert to Spikes per Second (Hz)

# 4. Plotting
plt.figure(figsize=(10, 5))
plt.plot(psth, color='navy', lw=2)
plt.fill_between(range(duration), psth, color='skyblue', alpha=0.3)
plt.title("PSTH: Average Social Neuron Response across 50 Patients")
plt.xlabel("Time (ms)")
plt.ylabel("Firing Rate (Hz)")
plt.grid(True, alpha=0.2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Create Synthetic Data (The "Ground Truth")
t = np.linspace(0, 10, 100).reshape(-1, 1)
intent = np.sin(t) # The Surgeon's actual emotional 'flow'

# 2. Simulate Neural Firing Rates (The 'Predictors')
# Neuron 1 follows the intent closely; Neuron 2 is noisy
neuron_1 = intent + np.random.normal(0, 0.1, (100, 1))
neuron_2 = np.random.normal(0, 0.5, (100, 1))

# 3. Train the Decoder (Linear Regression)
X = np.hstack([neuron_1, neuron_2]) # Combining the neurons into a matrix
model = LinearRegression().fit(X, intent)
reconstruction = model.predict(X)

# 4. Plotting the Result
plt.figure(figsize=(10, 5))
plt.plot(t, intent, label='Actual Surgeon Intent', color='black', lw=3, alpha=0.3)
plt.plot(t, reconstruction, label='Decoded from Brain Waves', color='red', linestyle='--')
plt.title("The 'Neural Decoder': Reconstructing Intent from Firing Rates")
plt.xlabel("Time")
plt.ylabel("Emotional Intensity")
plt.legend()
plt.show()

print(f"Decoder Weights: Clarity={model.coef_[0][0]:.2f}, Empathy={model.coef_[0][1]:.2f}")
```

```python
import numpy as np

# 1. Current State: [Clarity Weight, Empathy Weight]
weights = np.array([0.1, 0.9])
target_clarity = 0.5  # Our 'Equity' Goal
learning_rate = 0.1

print("--- Starting Weight Optimization ---")

# 2. The 'Learning' Loop (5 Iterations)
for i in range(1, 6):
    error = target_clarity - weights[0]
    # Nudge the weight closer to the target
    weights[0] += learning_rate * error
    # Balance it (so total weight stays roughly 1.0)
    weights[1] -= learning_rate * error

    print(f"Step {i}: Clarity={weights[0]:.2f}, Empathy={weights[1]:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.model_selection import train_test_split

# 1. Create Synthetic Population Data (100 Neurons, 200 'Moments')
n_neurons = 100
n_samples = 200

# Clear Comm: Neurons have a specific 'Success' pattern
clear_data = np.random.normal(0.5, 0.2, (100, n_neurons))
# Discordant Comm: Patterns are noisier and shifted
discordant_data = np.random.normal(0.7, 0.4, (100, n_neurons))

# Combine and Create Labels (0 = Clear, 1 = Discordant)
X = np.vstack([clear_data, discordant_data])
y = np.array([0]*100 + [1]*100)

# 2. Split into Training and Testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# 3. Train the SVM Classifier
clf = svm.SVC(kernel='linear')
clf.fit(X_train, y_train)

# 4. Check Accuracy
accuracy = clf.score(X_test, y_test)
print(f"Neural Decoder Accuracy: {accuracy * 100:.2f}%")
```

```python
# 1. Extract the 'Importance' (Weights) of each of the 100 neurons
neuron_importance = np.abs(clf.coef_[0])

# 2. Plot the 'Social Signature'
plt.figure(figsize=(12, 4))
plt.bar(range(100), neuron_importance, color='purple')
plt.title("Neuron 'Importance' in Detecting Language Discordance")
plt.xlabel("Neuron ID (0-99)")
plt.ylabel("Weight in the Decoder")
plt.show()

# 3. Find the 'Top Snitch'
top_neuron = np.argmax(neuron_importance)
print(f"Neuron #{top_neuron} is your 'Social Sentinel'. It is the most sensitive to discordance!")
```

```python
from sklearn.linear_model import LogisticRegression

# 1. Train a 'Sparse' Decoder using L1 (Lasso) penalty
# 'C' is the strength of our filter. Lower C = More Ruthless.
sparse_clf = LogisticRegression(penalty='l1', solver='liblinear', C=0.1)
sparse_clf.fit(X_train, y_train)

# 2. Count how many neurons are still 'Talking' (Weight > 0)
active_neurons = np.sum(sparse_clf.coef_ != 0)
print(f"The Lasso Filtered the noise! We are now decoding using only {active_neurons} neurons.")

# 3. Check Accuracy again
sparse_accuracy = sparse_clf.score(X_test, y_test)
print(f"Sparse Decoder Accuracy: {sparse_accuracy * 100:.2f}%")
```
