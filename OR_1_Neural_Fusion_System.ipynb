```python
import numpy as np

# ==========================================
# PHASE 7: NOISE FILTERING & RESILIENCE
# ==========================================

def deep_resilience_check(neural_signal):
    """
    Checks for hardware detachment or software freezes.
    Biological signals always have natural variance; a flat 0.0 variance implies a freeze.
    """
    if np.sum(neural_signal) == 0:
        return "HARDWARE_DISCONNECTED"
    if np.std(neural_signal) == 0.0:
        return "SOFTWARE_FREEZE_DETECTED"
    return "SIGNAL_LIVE"

def ultimate_guard(avg, volatility, bovie_active):
    """
    Filters out pure electrical noise (volatility > 100) and surgical tool startle
    to prevent alert fatigue in the operating theater.
    """
    if volatility > 100:
        return "SENSOR_ERROR"
    if avg > 110 and not bovie_active:
        return "ALARM"
    return "STABLE"

# ==========================================
# PHASE 8: ETHICAL ESCALATION & FEEDBACK
# ==========================================

def adaptive_check(current_val, threshold, bias):
    """Applies reinforcement learning adjustments based on human-in-the-loop overrides."""
    effective_threshold = threshold + bias
    if current_val > effective_threshold:
        return "ALARM"
    return "STABLE"

def ethical_alarm_routing(confidence, surgeon_dismissal):
    """Routes persistent high-confidence alarms to secondary staff if overridden."""
    if confidence > 0.90 and surgeon_dismissal:
        print("[ROUTING] Alerting ANESTHESIOLOGIST: High-Confidence Discordance Detected.")
        return "ALARM_ESCALATED"
    return "ALARM_SILENCED"

# ==========================================
# PHASE 9: THE MULTIMODAL GLM PIPELINE
# ==========================================

def run_or1_pipeline(neural_signal, vitals_signal, bovie_active, surgeon_dismissal=False, bias=0):
    """
    The main execution loop of the OR-1 Engine.
    Implements a deterministic Generalized Linear Model (GLM) for multimodal sensor fusion.
    """
    # 1. Verify Sensor Integrity
    status = deep_resilience_check(neural_signal)
    if status != "SIGNAL_LIVE":
        print(f"[WARNING] Operating in Degraded Mode (Vitals Only) due to: {status}")
        # Graceful degradation: Fall back entirely to vitals tracking
        w_neural, w_vitals = 0.0, 1.0
    else:
        # Standard Operating Procedure: 60/40 Neural-to-Metabolic Weighting
        w_neural, w_vitals = 0.6, 0.4

    avg_neural = np.mean(neural_signal)
    avg_vitals = np.mean(vitals_signal)
    volatility = np.std(vitals_signal)

    # 2. GLM Linear Predictor
    fused_score = (avg_neural * w_neural) + (avg_vitals * w_vitals)

    # 3. Decision Gate Processing
    result = ultimate_guard(fused_score, volatility, bovie_active)

    # 4. Handle Surgeon Overrides if the AI detects an emergency
    if result == "ALARM" and surgeon_dismissal:
        # Evaluate model confidence (simulated here for demonstration)
        simulated_confidence = 0.95
        result = ethical_alarm_routing(simulated_confidence, surgeon_dismissal)

    return result

# ==========================================
# RUNNING THE SIMULATION (Verification)
# ==========================================
if __name__ == "__main__":
    # Test Scenario: Acute distress with an stubborn surgeon override
    mock_brain_waves = [120, 115, 118, 122, 119] # High stress signature
    mock_heart_rate = [105, 110, 108, 112, 107]  # Elevated metabolic signature

    print("--- Initializing OR-1 Live Simulation ---")
    output = run_or1_pipeline(mock_brain_waves, mock_heart_rate, bovie_active=False, surgeon_dismissal=True)
    print(f"Final System State: {output}")
```
